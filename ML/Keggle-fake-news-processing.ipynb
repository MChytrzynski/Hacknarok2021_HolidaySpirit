```python
# imports
import pandas as pd
```

```python
# loading data from csv
test = pd.read_csv('../keggle-fake-news-dataset/test.csv')
train = pd.read_csv('../keggle-fake-news-dataset/train.csv')
```

```python
data =  pd.concat([train, test])
columns = ['id', 'title', 'author', 'content', 'veracityAI']
data.columns = columns
# df.rename(columns = {'Name' : 'First Name', 'age' : 'Age'}, inplace = True)
data.head()
```

```python
len(data.isna())
```

```python
data = data.dropna()
len(data.isna())
```

```python
data = data.sort_values(by=['veracityAI'])
```

```python
df = data.copy()
df
```

```python
for index, row in df.iterrows():
    if len(row['content']) < 2000 or len(row['content']) > 5000:
        data.drop(index, inplace=True)
    else: 
        if row['veracityAI'] == 1.0:
            data.at[index, 'veracityAI'] = 0
        else: 
            data.at[index, 'veracityAI'] = 1
```

```python
class source:
    id = ""
    name = ""
    def __init__(self, id, name):
        self.id = id
        self.name = name
```

```python
class article:
    source = None
    author = ""
    title = ""
    description = ""
    url = ""
    urlToImage = ""
    publishedAt = ""
    content = ""
    veracityAI = ""
    def __init__(self, source, author, title, description,url, urlToImage, publishedAt, content, veracityAI):
        self.source = source
        self.author = author
        self.title = title
        self.description = description
        self.url = url
        self.urlToImage = urlToImage
        self.publishedAt = publishedAt
        self.content = content
        self.veracityAI = veracityAI
```

```python
def obj_dict(obj):
    return obj.__dict__
```

```python
articles = []
for index, row in data.iterrows():
    src = source(str(row['id']), "")
    art = article(src, row['author'], row['title'], "", "", "", "", row['content'], row['veracityAI'])
    articles.append(art)
#     print(art.content)
#     if(index == 500):
#         break
```

```python
import json
fold = 200
first = 0
last = 0
length = len(articles)
for first in range(0, length, fold):
    if(first+fold > length):
        last = length-first-1
    else:
        last = first+fold
#     json_string = json.dumps(articles[first:last], default=obj_dict)
    json_string = json.dumps(articles[first:last], default=obj_dict, ensure_ascii=False).encode('utf8')
    json_string = json_string.decode().replace('\\n',"")
    json_string.replace('\\',"")
    json_string
    with open('jsons/articles_'+str(last)+'.json', 'w', encoding="utf-8") as f:
        f.write(json_string)
```
